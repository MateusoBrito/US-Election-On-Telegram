```python
vocab = set()
for message in df["text"].dropna():
    words = message.split()
    vocab.update(words)
print(f"Total unique words: {len(vocab)}")
```

# 1. Detecting Languages
---

```python
!wget https://dl.fbaipublicfiles.com/fasttext/supervised-models/lid.176.bin
!pip install fasttext-wheel
```

```python
import fasttext

MODEL_PATH = 'lid.176.bin'
model = fasttext.load_model(MODEL_PATH)

def detect_language_fasttext(text):
    text_clean = text.replace('\n', ' ')
    predict = model.predict(text_clean, k=1)
    language_label = predict[0][0]
    return language_label.replace('__label__', '')

df_limpo = df.dropna(subset=['text']).copy()
df_limpo['language'] = df_limpo['text'].apply(detect_language_fasttext)

display(df_limpo)
```

```python
count_language = df_limpo["language"].value_counts()
print(f"Number of Message per language:\n{count_language}")
```

```python
plt.figure(figsize=(10, 4))
count_language.head(10).plot(kind='bar', color='skyblue')

plt.title('Top 10 Most Common Languages in Messages')
plt.xlabel('Language')
plt.ylabel('Number of Messages')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
